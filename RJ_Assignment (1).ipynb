```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('2015.csv')
df
```

```python
#Duplicates
duplicatess = {}

for column in df.columns:
  duplicatess_cnt = df[column].duplicated().sum()
  duplicatess[column] = duplicatess_cnt

print("Number of duplicate values:")
print(duplicatess)
```

```python
#missing Values
no_of_missing_values= df.isna().sum()
no_of_missing_values
```

```python
#To get the number of columns and rows
df.shape
```

```python
#To understand column headings
df.columns
```

```python
#Unique_values-each variable.
for column in df.columns:
  unique_value=df[column].unique()
  print(f"The unique values, '{column}' is:")
  print(unique_value)
```

```python
#Begin pie chart
#calculating average
regional_happiness = df.groupby('Region')['Happiness Score'].mean()
print(regional_happiness)

# Plotting the Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(regional_happiness, labels=regional_happiness.index, autopct='%1.2f%%',
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
plt.title('Average Happiness Scores by Region',fontsize=14, pad=20)
plt.axis('equal')  # For making pie chart circular

plt.show()
```

```python
#Now lets plot Scatter plot
#Sorting Government corruption
Govt_Corruption_order= df.sort_values(by='Trust (Government Corruption)', ascending=False)
print(Govt_Corruption_order[['Country', 'Trust (Government Corruption)']])
#plotting Scatter plot
plt.scatter(df['Economy (GDP per Capita)'], df['Trust (Government Corruption)'])
plt.xlabel('Economy (GDP per Capita)',fontsize=12)
plt.grid(True)
plt.ylabel('Trust (Government Corruption)',fontsize=12)
plt.title('Economy vs. Trust',fontsize=14)
plt.show()
```

```python
#Begin Heat map.
#Selecting the required columns
Heading = ['Happiness Rank', 'Happiness Score',
       'Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
required_data=df[Heading]
```

```python
# Calculate correlation matrix
correlation_matrix = Heading.corr()
```

```python
# Create heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Heatmap')
plt.show()
```
